```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv('Walmart_sales.csv')
```

```python
df
```

```python
# Remove null values
df_clean = df.dropna()
```

```python

# Plot a histogram for Weekly Sales
plt.figure(figsize=(10, 6))
plt.hist(df_clean['Weekly_Sales'], bins=20, color='blue', edgecolor = "white")
plt.title('Distribution of Weekly Sales')
plt.xlabel('Weekly Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
# Select the first 100 rows of the dataset
new_df = df_clean.iloc[0:100,0:8]
```

```python
new_df
```

```python
# Creating a scatterplot graph
plt.figure(figsize=(10, 6))
plt.figure(dpi=144)
plt.scatter(new_df['Temperature'], new_df['Fuel_Price'])
plt.title('Relationship between Temperature and Fuel Price')
plt.xlabel('Temperature')
plt.ylabel('Fuel Price')
plt.show()
```

```python
# Exclude non-numeric column before correlation calculation
numeric_df = df.drop(['Date'], axis=1)

correlation_matrix = numeric_df.corr()

plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True,cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Pandas Describe Function
descriptive_metrics = new_df.describe()
descriptive_metrics
```
